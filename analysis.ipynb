```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
df_reviews = pd.read_excel("./chrono24_reviews.xlsx")
df_reviews = df_reviews[df_reviews['reviews year'].notna()]
df_reviews['reviews year'] = df_reviews['reviews year'].astype(int)
df_reviews.head()
```

```python
df_reviews.info()
```

```python
df_sellers = pd.read_excel("./chrono24_sellers.xlsx")
df_sellers = df_sellers[df_sellers['trusted since'].notna()]
df_sellers['trusted since'] = df_sellers['trusted since'].astype(int)
df_sellers.drop(columns=['other info'], inplace=True)
df_sellers.head()
```

```python
# Création du curseur pour sélectionner l'année
annee_selection = widgets.IntSlider(
  value=df_reviews['reviews year'].min(),
  min=df_reviews['reviews year'].min(),
  max=df_reviews['reviews year'].max(),
  step=1,
  description='Année:'
)

# Fonction de mise à jour du graphique
def update_barplot(annee):
  plt.figure(figsize=(8, 5))
  subset = df_reviews[df_reviews['reviews year'] == annee]
  sns.barplot(x='buyer_country', y='Valeur', data=subset, palette='Blues')
  plt.title(f'Données pour {annee}')
  plt.xlabel('Catégorie')
  plt.ylabel('Valeur')
  plt.show()

# Affichage interactif
widgets.interactive(update_barplot, annee=annee_selection)
```
