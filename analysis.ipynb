```python
import pandas as pd

# Load the data from the uploaded files
sellers_file_path = "chrono24_sellers.xlsx"
reviews_file_path = "chrono24_reviews.xlsx"

# Read the Excel files
sellers_df = pd.read_excel(sellers_file_path)
reviews_df = pd.read_excel(reviews_file_path)

# Display basic information about both datasets
sellers_info = sellers_df.info()
reviews_info = reviews_df.info()

# Display first few rows of both datasets
sellers_head = sellers_df.head()
reviews_head = reviews_df.head()

# Display summary statistics for numerical columns
sellers_desc = sellers_df.describe()
reviews_desc = reviews_df.describe()

# Display column names
sellers_columns = sellers_df.columns.tolist()
reviews_columns = reviews_df.columns.tolist()

# Show dataset sizes
sellers_shape = sellers_df.shape
reviews_shape = reviews_df.shape

sellers_info, reviews_info, sellers_head, reviews_head, sellers_desc, reviews_desc, sellers_columns, reviews_columns, sellers_shape, reviews_shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set(style="whitegrid")

# Plot distribution of GLOBAL RATING in sellers dataset
plt.figure(figsize=(8, 5))
sns.histplot(sellers_df["GLOBAL RATING"].dropna(), bins=20, kde=True)
plt.title("Distribution of Seller Global Ratings")
plt.xlabel("Global Rating")
plt.ylabel("Count")
plt.show()

# Countplot for sellers by country
plt.figure(figsize=(12, 6))
top_countries = sellers_df["country"].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Seller Countries")
plt.xlabel("Country")
plt.ylabel("Number of Sellers")
plt.xticks(rotation=45)
plt.show()


# Distribution of reviews per year in the reviews dataset
plt.figure(figsize=(8, 5))
sns.histplot(reviews_df["reviews year"].dropna(), bins=10, kde=True)
plt.title("Distribution of Reviews by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

# Count of reviews by buyer country (top 10)
plt.figure(figsize=(12, 6))
top_buyer_countries = reviews_df["buyer_country"].value_counts().head(10)
sns.barplot(x=top_buyer_countries.index, y=top_buyer_countries.values)
plt.title("Top 10 Buyer Countries by Number of Reviews")
plt.xlabel("Buyer Country")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45)
plt.show()
```
